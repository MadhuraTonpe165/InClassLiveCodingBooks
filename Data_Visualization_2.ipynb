```python
#Importing Important document
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Heatmap => coloured tabel
#creating the data

data = np.random.randint(low = 1,
                         high = 100,
                         size = (10, 10))
print('The data to be plotted :\n', data)
```

```python
# High value is never included means high value of 100 never come, only values till 99 will come
# colour need to be added in plural like blues, greens
```

```python
# Heatmap
sns.heatmap(data, annot = True, cmap = 'Blues')
#annot = True show the values in graph
```

```python
# Heatmap
sns.heatmap(data, annot = False, cmap = 'Blues')
#annot = False shows only colours
```

```python
#Application : Co-relation analysis, HeatMap is typically used for Co-Relation Analysis in ML Modeling
```

```python
#help
help(sns.heatmap)
help(sns.get_dataset_names)
sns.get_dataset_names #get names of all data set
```

```python
# CatPlot -> categorical Scatterplot
#Getting the data
exercise = sns.load_dataset('exercise', index_col = 0)
exercise.head()
```

```python
# catplot (continue---) time is categorical veriable and pluse is numerical
sns.catplot(x = 'time', y = 'pulse', data = exercise, hue = 'kind')
```

```python
# hue => separating the values by that column
```

```python
# distplot() => used to get the density curve along with histogram (kde value)
plt.figure(figsize = (5, 3))
sns.distplot(exercise['pulse'], kde = False, color = 'red')
plt.plot()
```

```python
sns.distplot(exercise['pulse'], kde = True, color = 'green')
# in kde d stands for density (Kernel density estimation) and pulse is continues random variable
```

```python
#Loading the 'tips' dataset
df = sns.load_dataset('tips')
df.head()
```

```python
# Pairplot = combination of every numericCAL VERIABLE WITH ALL OTHER NUMERICAL VERIABLES
# all diagonal elements will be univerient and all others are bivarient
sns.pairplot(df, hue = 'smoker', palette = 'coolwarm')
#
```

```python
'''
Pairplot is used to get bird's eye view of the dataset
Also in chart total_bill to Total_bill,distribution of data for smokers and non-smokers is almost same but there are few outliers in the data
'''
```

```python
# Jointplot()  - > Used for analyzing 2 variables along analysing a single variable a single variables as well
plt.figure(figsize = (5, 3))
sns.jointplot(x = 'total_bill', y = 'tip', data = df)
plt.show()
```

```python
plt.figure(figsize = (5, 3))
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'kde') # default value for kind is scatter
plt.show()
```

```python
#if a region is surrounded by multiple circles then density is high
```

```python
plt.figure(figsize = (5, 3))
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hex')
plt.show()
```

```python
# darker the colour higher is density, meand high datapoints
```

```python
plt.figure(figsize = (5, 3))
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg')
plt.show()
```

```python
# shaded region is 95% confidence interval , line is regression
```

```python
'''
63kind = scatter, kde, hex, reg
are different types of joint plots
'''
```
