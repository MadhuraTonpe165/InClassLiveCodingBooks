```python
# Creating a DataFrame (self practise)

df1 = pd.DataFrame({
    'Key' : ['K0', 'K1', 'K2', 'K3'],
    'Key1' : ['K0', 'K1', 'K0', 'K1'],
    'Name' : ['Abhay', 'Bobby', 'Chandan', 'Danny'],
    'Age' : [25, 28, 31, 40]
})

df2 = pd.DataFrame({
    'Key' : ['K0', 'K1', 'K2', 'K3'],
    'Key1' : ['K0', 'K0', 'K0', 'K0'],
    'Address' : ['Pune', 'Ahmadabad', 'Nagpur', 'Delhi'],
    'Qualification' : ['BA','BBA','BCA','MCA']
})

print(df1, '\n\n',  df2)
```

```python
# # (k0,k0) and (k2,k0) are the only common combinations
```

```python

# merge with multiple keys(composite key)
res = pd.merge(df1, df2, on = ['Key', 'Key1'])
print(res)
```

```python
# Merge wit multiple keys (Composite keys)
res = pd.merge(df1, df2, how = 'left', on = ['Key', 'Key1'])
print(res)
```

```python
# the merger operation will take all columns from first(left) dataset and add NaN as values where same combination is not available in right dataset
```

```python
# Merge wit multiple keys (Composite keys)
res = pd.merge(df1, df2, how = 'right', on = ['Key', 'Key1'])
print(res)
```

```python
# Value of how will decide witch join will be applied
# if any one value is nan the int dtype will be converted to float
```

```python
# Merge wit multiple keys (Composite keys)
res = pd.merge(df1, df2, how = 'outer', on = ['Key', 'Key1'])
print(res)
```

```python
# Outer all combinations will come for inner only common combinations will come. Inner is default value for parameter how in merger function
```

```python
# Merge wit multiple keys (Composite keys)
res = pd.merge(df1, df2, how = 'inner', on = ['Key', 'Key1'])
print(res)
```

```python
# Creating a DataFrame

df = pd.DataFrame({
    'Product' : ['Carrot', 'Broccoli', 'Banana', 'Banana', 'Beans','Orange', 'Broccoli','Banana'],
    'Category' : ['Vegetable','Vegetable','Fruit','Fruit','Vegetable','Fruit','Vegetable','Fruit'],
    'Quantity' : [8,5,3,4,6,4,7,8],
    'Amount' : [270,240,620,384,626,610,620,190]
})

df.head()
```

```python
# Pivot Table with sum aggregation

pivot = df.pivot_table(index = ['Product'],
                       values = ['Amount'],
                       aggfunc = 'sum'
                       )
print(pivot)
```

```python
# Pivot Table with sum aggregation

pivot = df.pivot_table(index = ['Category'],
                       values = ['Amount'],
                       aggfunc = 'sum'
                       )
print(pivot)
```

```python
# Pivot Table with sum aggregation

pivot = df.pivot_table(index = ['Product' , 'Category'],
                       values = ['Amount'],
                       aggfunc = 'sum')
print(pivot)
```

```python
# Pivot Table with sum aggregation

pivot = df.pivot_table(index = ['Product' , 'Category'],
                       values = ['Amount', 'Quantity'],
                       aggfunc = 'sum')
print(pivot)
```

```python
# Pivot Table with sum aggregation

pivot = df.pivot_table(index = ['Product' , 'Category'],
                       values = ['Amount', 'Quantity'],
                       aggfunc = ['sum', 'median'])
print(pivot)
```
