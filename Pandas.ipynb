```python
# Creatin the Dataset
data1 = {
    'Name' : ['Pranav', 'Sai', 'Niraj', 'Vijay'],
    'Age' : [27, 28, 29, 30],
}

data2 = {
    'Address' : ['Mumbai', 'Delhi', 'Wardha', 'Vizag'],
    'Qualification' : ['BE', 'BTech', 'MSc', 'MA']
}

# Converting to dataframes
df = pd.DataFrame(data1, index = [0, 1, 2, 3])
df1 = pd.DataFrame(data2, index = [2, 3, 6, 7])

print(df, '\n\n', df1)
```

```python
# Column Concatination
res = pd.concat([df, df1], axis = 1)
print(res)
```

```python
'''
axis parameter has only 2 values -> 0 and 1
Default value for axis is 0 -> row wise Concatination
if we manually set axis to 1 -> column wise peration
'''
```

```python
#inner join only similer index from  both dataset
res = pd.concat([df, df1], axis = 1, join = 'inner')
print(res)
```

```python
# by default join is set o outer
res = pd.concat([df, df1], axis = 1, join = 'outer')
print(res)
```
