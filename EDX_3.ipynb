```python
'''
Question 5 - >
Perform correlation analysis on the data and comment on it.
'''
```

```python
# Solution ->

plt.figure(figsize = (10,8))
sns.heatmap(df.corr(numeric_only= True), annot = True)
plt.title('HeatMap')
plt.show()
```

```python
'''
Conclusions - >
(1) "year" and "milage" have a strong negative correlation
(2) "milgae and "Price" have a moderate negative correlation
(3) "mpg"  and "tax" have a negative correlation
($) "mpg" and "price"
'''
```

```python
'''
Question 6 - >

Plot the density curve for "tax"
'''
```

```python
# Solution - >

plt.figure(figsize = (8,5))
sns.kdeplot(df["tax"] )
plt.title('Density Curve')
plt.show()
```

```python
# The data looks bimodel with some outliers in the left tail
```

```python
'''
Question 7 ->
Plot a boxplot for "year" and "price"
'''
```

```python
# Solution - >

plt.figure(figsize=(15, 6))
sns.boxplot(x = df['year'], y = df['price'], data = df, palette = 'Set1')
plt.title('Boxplot')
plt.show()
```

```python
'''
Conclusion - >

(1) Year 1970 seems to be the outlier
(2) From year 1997 till 2010, the data seems to be in a constant range
(3) After 2010, there has been a linear increase in the price till 2020
(4) the number of outliers in the data from 2014 have severly increased
'''
```
